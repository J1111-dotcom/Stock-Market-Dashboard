# Stock Market Trend Analysis (Notebook Version)
This notebook analyzes historical stock performance for Tesla, Amazon, AMD and Gamestop using Python and the yFinance API.

## Use Case
To identify trends and patterns in stock prices to support potential investment decisions.

```python
#Importing relevant libraries

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

#Enable inline plotting for Jupyter
%matplotlib inline
```

```python
#Defining stock tickers

tickers = {
    'Tesla' : 'TSLA',
    'Amazon' : 'AMZN',
    'AMD' : 'AMD',
    'GameStop' : 'GME'
}
```

```python
stock_data = {}
# Downloading historical data

for name, symbol in tickers.items():
    data = yf.Ticker(symbol).history(period="5y")
    data.reset_index(inplace=True)
    stock_data[name] = data
```

## Data Preview

Below is a quick preview of the downloaded data for each stock, showing the first few rows of Open and Close prices.


```python
for name, df in stock_data.items():
    print(f"{name} stock data:")
    print(df[['Date', 'Open', 'Close']].head(3))  # Show just the first 3 rows
    print("\n")
```

```python
#Plotting Open/Close Prices Over Time

for name, data in stock_data.items():
    plt.figure(figsize=(12,5))
    plt.plot(data['Date'], data['Open'], label='Open Price')
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(f"{name} Stock Price (Open vs Close)")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Key Insights

**Tesla:**  
- A notable upward trend between 2020 and 2021 with subsequent high volatility.  
- This suggests strong growth periods followed by market corrections.

**Amazon:**  
- Exhibits steady growth with relatively low volatility, indicating resilient performance over the analyzed period.

**AMD:**  
- Demonstrates significant growth, likely fueled by the increasing demand in the semiconductor market.

**GameStop:**  
- Shows a dramatic price spike in early 2021, reflecting the impact of retail trading dynamics and market hype.
